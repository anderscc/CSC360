```python
import queue
import copy
import random
```

```python
def FIFO(arr, frameSize):
    arrCopy = copy.deepcopy(arr)
    arr1 = []
    counter = 0
    while(len(arrCopy)!=0):
        for i in arr:
            if (len(arr1) < frameSize):
                arr1.append(i)
                counter=counter +1
                arrCopy.pop(0)

            elif (len(arr1) >= frameSize and i not in arr1):
                arr1.pop(0)
                arr1.append(i)
                arrCopy.pop(0)
                counter = counter +1

            else:
                arrCopy.pop(0)
    return counter


        
```

```python
def RAND(arr, frameSize):
    arrCopy = copy.deepcopy(arr)
    arr1 = []
    counter = 0
    while(len(arrCopy)!=0):
        for i in arr:
            if (len(arr1) < frameSize):
                arr1.append(i)
                counter=counter +1
                arrCopy.pop(0)

            elif (len(arr1) >= frameSize and i not in arr1):
                randomNumber = random.randint(0,frameSize-1)
                arr1.pop(randomNumber)
                arr1.append(i)
                arrCopy.pop(0)
                counter = counter +1

            else:
                arrCopy.pop(0)
    return counter
```

```python
def LRU(arr, frameSize):
    arrCopy = copy.deepcopy(arr)
    arr1 = []
    counter = 0
    while(len(arrCopy)!=0):
        for i in arr:
            if (len(arr1) < frameSize):
                arr1.append(i)
                counter=counter +1
                arrCopy.pop(0)

            elif (len(arr1) >= frameSize and i not in arr1):
                arr1.pop(0)
                arr1.append(i)
                arrCopy.pop(0)
                counter = counter +1

            elif (len(arr1) >= frameSize and i in arr1):
                arr1.remove(i)
                arr1.append(i)
                arrCopy.pop(0)

            else:
                arrCopy.pop(0)
    return counter
```
